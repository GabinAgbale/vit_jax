```python
import os
import sys

import jax
import jax.numpy as jnp
from jax import random 

from PIL import Image
import matplotlib.pyplot as plt

parent_dir = os.getcwd()
data_dir = os.path.join(parent_dir, "..")
sys.path.append(data_dir)

from data import patcher

main_rng = random.PRNGKey(42)
```

```python
path_monkey = "/Users/gabin/PycharmProjects/jax_ViT/monkey.jpeg"
img = jnp.asarray(Image.open(path_monkey))
plt.imshow(img)
img = img.reshape(1, img.shape[0], img.shape[1], img.shape[2])
print(img.shape)
```

```python
p = patcher.Patcher(20, 24)
patches = p._create_patches(img)
print(patches.shape)
```

```python
plt.imshow(patches[0, 10, :, :, :])
```

```python
main_rng, init_rng = random.split(main_rng, 2)
params = p.init({"params": init_rng}, img)['params']
# Apply encoder block with parameters on the inputs
# Since dropout is stochastic, we need to pass a rng to the forward
main_rng, dropout_apply_rng = random.split(main_rng)
out = p.apply({'params': params}, img)
print('Out', out.shape)

del p, params
```
